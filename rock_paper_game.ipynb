```python
import random

user_score = 0
computer_score = 0

while True:
    print("\n1. Rock")
    print("2. Paper")
    print("3. Scissors")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 4:
        break

    if choice == 1:
        user = "Rock"
    elif choice == 2:
        user = "Paper"
    elif choice == 3:
        user = "Scissors"
    else:
        print("Invalid Choice")
        continue

    computer = random.randint(1, 3)

    if computer == 1:
        comp = "Rock"
    elif computer == 2:
        comp = "Paper"
    else:
        comp = "Scissors"

    print("You chose:", user)
    print("Computer chose:", comp)

    if user == comp:
        print("Match Tie")
    elif (user == "Rock" and comp == "Scissors") or \
         (user == "Paper" and comp == "Rock") or \
         (user == "Scissors" and comp == "Paper"):
        print("You Win")
        user_score += 1
    else:
        print("Computer Wins")
        computer_score += 1

    print("Your Score:", user_score)
    print("Computer Score:", computer_score)

print("\nFinal Score")
print("Your Score:", user_score)
print("Computer Score:", computer_score)
```
