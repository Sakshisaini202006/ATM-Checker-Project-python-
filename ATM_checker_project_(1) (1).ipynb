```python
username = "Sakshi"
password = "1234"

while True:
    user = input("Enter Username: ")

    if user != username:
        print("Invalid Username")
    else:
        pwd = input("Enter Password: ")

        if pwd != password:
            print("Invalid Password")
        else:
            print("Login Successful")
            break

balance = 5000

while True:
    print("1 Deposit")
    print("2 Withdraw")
    print("3 Balance")
    print("4 Exit")

    option = int(input("Enter your option: "))

    if option == 1:
        amt = int(input("Enter the amount: "))
        balance += amt

    elif option == 2:
        amt = int(input("Enter the amount: "))
        balance -= amt

    elif option == 3:
        print("Balance =", balance)

    elif option == 4:
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
